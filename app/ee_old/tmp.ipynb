```python
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
from torchvision.utils import make_grid
import random

# データセットのダウンロードと変換
transform = transforms.Compose([transforms.ToTensor()])
trainset = torchvision.datasets.CIFAR10(root='/home/haselab/Documents/tat/Research/assets/datasets', train=True, download=True, transform=transform)

# クラスラベル
classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')

# 各クラスの画像を取得
images_per_class = {class_label: [] for class_label in classes}
for image, label in trainset:
    class_label = classes[label]
    images_per_class[class_label].append(image)

# 各クラスからランダムに10枚の画像を選択
random_images_per_class = {class_label: random.sample(images, 10) for class_label, images in images_per_class.items()}

# 画像を縦に結合して表示
fig, axs = plt.subplots(10, 10, figsize=(15,15))
for i, class_label in enumerate(classes):
    for j, img in enumerate(random_images_per_class[class_label]):
        npimg = img.numpy()
        axs[i, j].imshow(np.transpose(npimg, (1, 2, 0)))
        axs[i, j].axis('off')
        if j == 0:  # 最初の列のみクラス名を表示
            axs[i, j].text(-0.2, 0.5, class_label + " :", size=28, ha='right', va='center', color='black', transform=axs[i, j].transAxes)
  
fig.show()
```
